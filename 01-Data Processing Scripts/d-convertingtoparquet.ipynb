# Converting CSV dataset to Parquet

```python
import pandas as pd
```

```python
# Change these filenames as needed
input_file = "/Users/pdeguz01/Documents/git/IDS705_TeamPandas/LargeData/restauranthealthinspections2024_CensusSVIUSDA_halfmileradius.csv"
output_file = "/Users/pdeguz01/Documents/git/IDS705_TeamPandas/00-Data/Data Viz Data/MergedDataset.parquet"

# Read CSV exactly as-is (preserves structure)
df = pd.read_csv(input_file, dtype=str)

# Write to Parquet
df.to_parquet(output_file, index=False, compression="snappy")

print("✅ Done! Saved as", output_file)
```

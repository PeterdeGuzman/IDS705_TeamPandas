```python
raw_lac = pd.read_csv(
    "/Users/jw/Desktop/MIDS SP25/Machine Learning/Final Project/Restaurant_and_Market_Health_Inspections.csv",
    encoding="ISO-8859-1",
)
raw_lac
```

```python
# view cols
print(raw_lac.columns.tolist())
```

```python
# drop unneeded columns
lac_colstokeep = [
    "activity_date",
    "facility_name",
    "facility_address",
    "facility_city",
    "facility_zip",
    "pe_description",
    "service_description",
    "score",
    "grade",
    "serial_number",
]
lac_fewercols = raw_lac[lac_colstokeep]
```

```python
# Clean dates
# convert "INSPECTION_DATE" to datetime
lac_fewercols["activity_date"] = pd.to_datetime(
    lac_fewercols["activity_date"].str.strip(), format="%m/%d/%Y %H:%M:%S %p"
)
# Creating new column with Year of Inspection Date
lac_fewercols.loc[:, "INSPDATE_YEAR"] = lac_fewercols["activity_date"].dt.year
```
